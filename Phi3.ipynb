```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the model
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",  # Automatically use GPU if available
    torch_dtype="auto",  # Automatically select the best precision
    trust_remote_code=True  # Allow custom model code
)

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline

# Create a pipeline for text generation
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,  # Exclude the input prompt in the output
    max_new_tokens=500,  # Limit output to 500 tokens
    do_sample=False  # Use deterministic text generation
)
```

```python
# Example prompt
prompt = "Tell me the capital of France"

# Generate text
output = generator(prompt)

# Print the generated text
print(output[0]["generated_text"])
```

```python
# Step 1: Install Required Libraries
!pip install transformers accelerate

# Step 2: Import Libraries
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Step 3: Check GPU Availability
if torch.cuda.is_available():
    print("Using GPU:", torch.cuda.get_device_name(0))
else:
    print("Using CPU")

# Step 4: Load the Model and Tokenizer
# Ensure model is loaded onto the available device (GPU or CPU)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",  # Automatically use GPU if available
    torch_dtype="auto",  # Automatically adjust precision
    trust_remote_code=True  # Use custom implementation for this model
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Step 5: Create the Text-Generation Pipeline
# Configure the generator pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=True,  # Exclude the input prompt in the output
    max_new_tokens=200,  # Limit output to 100 tokens (adjustable)
    do_sample=False  # Use deterministic output (greedy decoding)
)

# Step 6: Define a Prompt and Generate Text
prompt = "Explain the significance of artificial intelligence in healthcare:"
output = generator(prompt)

# Step 7: Print the Output
print("Generated Text:\n", output[0]["generated_text"])
```
